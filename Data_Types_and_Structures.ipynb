# Theoratical questions

1. What are data structures, and why are they important?
  - Data structures are ways of organizing and storing data so that they can be accessed and modified efficiently. They are important because they help manage large amounts of data, improve performance, and make programs more efficient and easier to understand.

2. Explain the difference between mutable and immutable data types with examples?
  - Mutable data types can be changed after creation (e.g., list, dictionary, set).
Immutable data types cannot be changed once created (e.g., string, tuple, int).

3. What are the main differences between lists and tuples in Python?
  - Lists are mutable, meaning their elements can be modified, while tuples are immutable.
Lists use square brackets [], and tuples use parentheses (). Lists are generally slower but more flexible than tuples.

4. Describe how dictionaries store data?
  - Dictionaries store data as key-value pairs. Each key is unique, and Python uses a hashing mechanism to quickly retrieve the value associated with a key.

5.  Why might you use a set instead of a list in Python?

  - You use a set when you need unique elements and don’t care about order. Sets also provide faster membership testing compared to lists.

6. What is a string in Python, and how is it different from a list?
  -  A string is a sequence of characters, while a list is a collection of any data types. Strings are immutable, whereas lists are mutable.

7. How do tuples ensure data integrity in Python?
  - Tuples are immutable, meaning their contents cannot be changed after creation. This immutability ensures that data remains consistent and protected from accidental modification.  

8. What is a hash table, and how does it relate to dictionaries in Python?
  - A hash table stores data using a hash function to map keys to values for quick access. Python dictionaries are implemented internally using hash tables.

9. Can lists contain different data types in Python?
  - Yes, lists can contain elements of different data types, such as integers, strings, and even other lists.

10. Explain why strings are immutable in Python.
  - Strings are immutable to ensure data consistency, security, and efficiency. Once created, a string’s value cannot change, preventing accidental modification.

11. What advantages do dictionaries offer over lists for certain tasks?
  - Dictionaries provide fast lookups using keys, making them ideal for scenarios where you need to quickly find or update data based on a specific identifier.

12. Describe a scenario where using a tuple would be preferable over a list.
  - Tuples are preferred when you want to store fixed data that shouldn’t change, such as coordinates (x, y) or configuration settings.  

13. How do sets handle duplicate values in Python?
  - Sets automatically remove duplicates . only unique elements are stored, regardless of how many times they’re added.

14. How does the “in” keyword work differently for lists and dictionaries?
  - In a list, in checks whether a value exists among the elements.
In a dictionary, in checks whether a key exists, not a value.

15. Can you modify the elements of a tuple? Explain why or why not.
  - No, you cannot modify a tuple’s elements because tuples are immutable  ,once created, their content cannot be changed or updated.  

16. What is a nested dictionary, and give an example of its use case.
  - A nested dictionary is a dictionary within another dictionary.  
    ex: students = {"John": {"age": 20, "grade": "A"}}

17. Describe the time complexity of accessing elements in a dictionary.
  - Accessing elements in a dictionary has an average time complexity of O(1) due to hashing, meaning lookups are very fast.

18. In what situations are lists preferred over dictionaries?
  - Lists are preferred when the order of elements matters, or when you only need to store data without associating keys to values.

19. Why are dictionaries considered unordered, and how does that affect data retrieval?
  - Dictionaries are considered unordered because their elements are stored based on hash values of keys. You retrieve data using keys, not positions.

20. Explain the difference between a list and a dictionary in terms of data retrieval ?
  - Lists use index positions to retrieve elements, meaning you access data by its order.
Dictionaries, on the other hand, use unique keys to retrieve values, allowing faster and more meaningful access.
Lists are better for ordered collections, while dictionaries are ideal for key-based lookups.
