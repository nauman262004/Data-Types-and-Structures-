# Data ToolKit Theory Questions

1. What is NumPy, and why is it widely used in Python?

  - NumPy is a Python library for numerical computation. It is widely used because it provides fast arrays, mathematical functions, and efficient operations for large datasets.

2. How does broadcasting work in NumPy?

  - Broadcasting allows NumPy to perform operations on arrays of different shapes by automatically expanding them when possible. This makes calculations simpler and faster.

3. What is a Pandas DataFrame?

  - A DataFrame is a 2D table-like data structure in Pandas with rows and columns. It is used for storing, analyzing, and manipulating structured data.

4. Explain the use of the groupby() method in Pandas.

  - groupby() groups data based on one or more columns. It helps perform operations like sum, mean, or count on each group separately.

5. Why is Seaborn preferred for statistical visualizations?

  - Seaborn provides beautiful, high-level, and easy-to-use plots. It is built on Matplotlib and is great for visualizing statistical patterns in data.

6. What are the differences between NumPy arrays and Python lists?

  - NumPy arrays are faster, use less memory, and support mathematical operations. Python lists are slower and meant for general-purpose data storage.

7. What is a heatmap, and when should it be used?

  - A heatmap is a graphical representation of data where values are shown using colors. It is used to visualize correlations, patterns, or density in a dataset.

8. What does the term “vectorized operation” mean in NumPy?

  - Vectorized operations allow you to perform calculations on entire arrays at once, without loops. This makes the code fast and efficient.

9. How does Matplotlib differ from Plotly?

  - Matplotlib is static and great for simple scientific plots. Plotly is interactive, allowing zooming, hovering, and dynamic visualizations.

10. What is the significance of hierarchical indexing in Pandas?

  - Hierarchical indexing allows multiple levels of indexes in rows or columns. It helps work with complex, multi-dimensional data more easily.

11. What is the role of Seaborn’s pairplot() function?

  - pairplot() creates multiple plots showing relationships between all pairs of numerical features. It is useful for understanding patterns in datasets.

12. What is the purpose of the describe() function in Pandas?

  - describe() provides summary statistics like mean, count, min, max, and percentiles. It helps quickly understand data distribution.

13. Why is handling missing data important in Pandas?

  - Missing data can affect calculations, charts, and model accuracy. Pandas provides tools to detect, fill, or remove missing values.

14. What are the benefits of using Plotly for data visualization?

  - Plotly creates interactive, web-based, and highly customizable visualizations. It is ideal for dashboards and presentations.

15. How does NumPy handle multidimensional arrays?

  - NumPy uses ndarray objects that can represent data in multiple dimensions. It provides tools to reshape, index, slice, and perform operations across dimensions.  

16. What is the role of Bokeh in data visualization?

  - Bokeh is used to create interactive, web-friendly visualizations in Python. It allows zooming, panning, and live data updates directly in the browser.

17. Explain the difference between apply() and map() in Pandas?

  - map() works on a single Series to transform each value.
apply() works on both Series and DataFrames and can apply complex functions across rows or columns.

18. What are some advanced features of NumPy?

  - NumPy offers features like broadcasting, vectorization, linear algebra, random number generation, and multi-dimensional array manipulation. These make it powerful for scientific computing.  

19. How does Pandas simplify time series analysis?

  - Pandas provides built-in date/time indexing, resampling, shifting, rolling windows, and frequency conversion. These tools make handling time-based data easy and efficient.

20. What is the role of a pivot table in Pandas?

  - A pivot table summarizes data by reorganizing it using rows, columns, and aggregated values. It is useful for quick data analysis and comparisons.

21. Why is NumPy’s array slicing faster than Python’s list slicing?

  - NumPy arrays use continuous memory blocks and optimized C code. Python lists store objects separately, making slicing slower.

22. What are some common use cases for Seaborn?

  - Seaborn is used for visualizing distributions, correlations, categorical data, heatmaps, and regression relationships. It makes statistical plots simple and attractive.  

# Practical Questions

```python
# 1. How do you create a 2D NumPy array and calculate the sum of each row?

import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])
sum = arr.sum(axis=1)
print(sum)
```

```python
# 2. Write a Pandas script to find the mean of a specific column in a DataFrame.

import pandas as pd

df = pd.DataFrame({'Marks': [80, 90, 75, 85]})
print(df['Marks'].mean())
```

```python
# 3. Create a scatter plot using Matplotlib.

import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 15, 20, 25]
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.scatter(x, y)
plt.show()
```

```python
# 4. Calculate correlation matrix using Seaborn and visualize with a heatmap.

import seaborn as sns

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
corr = df.corr()
sns.heatmap(corr, annot=True)
plt.show()
```

```python
# 5. Generate a bar plot using Plotly.

import plotly.express as px

df = pd.DataFrame({'Category': ['A', 'B', 'C'], 'Values': [10, 15, 20]})
fig = px.bar(df, x='Category', y='Values')
fig.show()
```

```python
# 6. Create a DataFrame and add a new column based on an existing column.

df = pd.DataFrame({'A': [1, 2, 3]})
df['B'] = df['A'] * 2
(df)
```

```python
# 7. Write a program to perform element-wise multiplication of two NumPy arrays.

arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
result = arr1 * arr2
print(result)
```

```python
# 8. Create a line plot with multiple lines using Matplotlib.

import matplotlib.pyplot as plt

x = [1,2,3,4]
y1 = [2,4,6,8]
y2 = [3,6,9,12]
plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.legend()
plt.title('Multiple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.plot(x, [1,4,9,16])
plt.plot(x, [2,3,2,3])
plt.show()

```

```python
# 9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.

df = pd.DataFrame({'Marks': [80, 90, 75, 85]})
threshold = 80
filtered_df = df[df['Marks'] > threshold]
(filtered_df)
```

```python
# 10. Create a histogram using Seaborn to visualize a distribution.

import seaborn as sns

df = pd.DataFrame({'Marks': [80, 90, 75, 85]})
plt.title('Histogram')
plt.xlabel('Marks')
plt.ylabel('Frequency')
sns.histplot(df['Marks'])
plt.show()
```

```python
# 11. Perform matrix multiplication using NumPy.

arr1 = np.array([[12, 10], [13, 24]])
arr2 = np.array([[5, 15], [37, 18]])
result = np.dot(arr1, arr2)
print(result)
```
