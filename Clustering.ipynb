# Clustering Theory Questions

1. What is unsupervised learning in the context of machine learning?
 - Unsupervised learning is a type of machine learning where the model is trained on data without labeled outputs. The algorithm tries to discover hidden patterns or structures in the data. It is commonly used for clustering, association, and dimensionality reduction.

2. How does K-Means clustering algorithm work?
 - K-Means groups data into K clusters based on similarity. It starts by randomly selecting K centroids and assigns each data point to the nearest centroid. The centroids are then recalculated based on the cluster points. This process repeats until the centroids no longer change.

3. Explain the concept of a dendrogram in hierarchical clustering.
 - A dendrogram is a tree-like diagram used to show the arrangement of clusters in hierarchical clustering. It illustrates how clusters are merged

4. What is the main difference between K-Means and Hierarchical Clustering?
 - K-Means requires the number of clusters (K) to be specified in advance, while hierarchical clustering does not. Hierarchical clustering builds a tree of clusters (dendrogram). K-Means is usually faster for large datasets, while hierarchical clustering is more detailed.

5. What are the advantages of DBSCAN over K-Means?
  - DBSCAN can detect clusters of arbitrary shapes, while K-Means works best with spherical clusters. It can also identify noise and outliers automatically. Unlike K-Means, DBSCAN does not require specifying the number of clusters beforehand.

6. When would you use Silhouette Score in clustering?
  - Silhouette Score is used to evaluate the quality of clustering. It measures how similar a data point is to its own cluster compared to other clusters. It helps in determining the optimal number of clusters and validating clustering results.

7. What are the limitations of Hierarchical Clustering?
 - Hierarchical clustering is computationally expensive for large datasets. Once clusters are merged or split, the process cannot be reversed. It is also sensitive to noise and outliers, which may affect the clustering structure.

8. Why is feature scaling important in clustering algorithms like K-Means?
 - K-Means uses distance calculations (usually Euclidean distance) to form clusters. If features have different scales, variables with larger values dominate the distance calculation. Feature scaling ensures all variables contribute equally to clustering.

9. How does DBSCAN identify noise points?
  - DBSCAN classifies points based on density of neighboring points. If a point does not have enough neighbors within a defined radius, it is labeled as noise or an outlier. These points are not assigned to any cluster.

10. Define inertia in the context of K-Means.
  - Inertia measures the sum of squared distances between data points and their cluster centroid. It represents how tightly the data points are grouped within clusters. Lower inertia indicates better and more compact clustering.

11. What is the elbow method in K-Means clustering?
  - The elbow method helps determine the optimal number of clusters (K). It plots the inertia against different values of K. The point where the decrease in inertia starts to slow down forms an “elbow,” indicating the best K value.

12. Describe the concept of "density" in DBSCAN.
  - In DBSCAN, density refers to the number of data points within a specified radius (epsilon) around a point. Regions with many nearby points are considered dense areas and form clusters. Sparse regions are treated as noise or boundaries.  

13. Can hierarchical clustering be used on categorical data?
  - Yes, hierarchical clustering can be applied to categorical data using appropriate distance measures such as Hamming distance or similarity metrics. It depends on the choice of distance or similarity measure rather than the data type itself.

14. What does a negative Silhouette Score indicate?
  - A negative Silhouette Score means that a data point is closer to another cluster than to its own cluster. This suggests that the point may be misclassified or poorly clustered. It indicates bad clustering structure.

15. Explain the term "linkage criteria" in hierarchical clustering.
  - Linkage criteria determine how the distance between clusters is calculated during hierarchical clustering. Examples include single linkage, complete linkage, and average linkage. The choice of linkage affects how clusters are formed.  

16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?
  - K-Means assumes clusters are spherical and of similar size and density. When clusters have different shapes or densities, K-Means may incorrectly assign points. This can result in poor clustering accuracy.

17. What are the core parameters in DBSCAN, and how do they influence clustering?
  - The main parameters are epsilon (ε) and minPts. Epsilon defines the radius for neighborhood search, while minPts specifies the minimum number of points required to form a dense region. These parameters control cluster formation and noise detection.

18. How does K-Means++ improve upon standard K-Means initialization?
  - K-Means++ improves centroid initialization by selecting initial centroids more strategically. It spreads the centroids far apart before clustering begins. This helps reduce poor clustering results and speeds up convergence.  

19. What is agglomerative clustering?
  - Agglomerative clustering is a bottom-up hierarchical clustering method. Each data point starts as its own cluster. Clusters are then gradually merged based on similarity until all points form a single cluster.

20. What makes Silhouette Score a better metric than just inertia for model evaluation?
  - Inertia only measures compactness within clusters, but it does not consider separation between clusters. Silhouette Score evaluates both cohesion (within cluster) and separation (between clusters). Therefore, it provides a more balanced measure of clustering quality.  

# Clustering Practical Questions

```python
# 21. Generate synthetic data with 4 centers using make_blobs and apply K-Means

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples=300, centers=4, random_state=42)

kmeans = KMeans(n_clusters=4)
labels = kmeans.fit_predict(X)

plt.scatter(X[:,0], X[:,1], c=labels)
plt.title("K-Means Clustering with 4 Centers")
plt.show()
```

```python
# 22. Load Iris dataset and use Agglomerative Clustering

from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering

iris = load_iris()
X = iris.data

model = AgglomerativeClustering(n_clusters=3)
labels = model.fit_predict(X)

print(labels[:10])
```

```python
# 23. Generate data using make_moons and apply DBSCAN

from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

X, y = make_moons(n_samples=300, noise=0.05)

db = DBSCAN(eps=0.2, min_samples=5)
labels = db.fit_predict(X)

plt.scatter(X[:,0], X[:,1], c=labels)
plt.title("DBSCAN with Outliers")
plt.show()
```

```python
# 24. Load Wine dataset and apply K-Means after standardization

from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import numpy as np

wine = load_wine()
X = wine.data

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=3)
labels = kmeans.fit_predict(X_scaled)

print("Cluster sizes:", np.bincount(labels))
```

```python
# 28. Digits dataset → PCA → K-Means visualization

from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

digits = load_digits()
X = digits.data

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

kmeans = KMeans(n_clusters=10)
labels = kmeans.fit_predict(X_pca)

plt.scatter(X_pca[:,0], X_pca[:,1], c=labels)
plt.title("K-Means on Digits (PCA)")
plt.show()
```

```python
# 29. Silhouette scores for k = 2 to 5

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples=300, centers=4)

scores = []
k_values = range(2,6)

for k in k_values:
    kmeans = KMeans(n_clusters=k)
    labels = kmeans.fit_predict(X)
    scores.append(silhouette_score(X, labels))

plt.bar(k_values, scores)
plt.xlabel("K")
plt.ylabel("Silhouette Score")
plt.show()
```

```python
# 30. Hierarchical clustering dendrogram

from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

iris = load_iris()
X = iris.data

Z = linkage(X, method='average')

dendrogram(Z)
plt.title("Hierarchical Clustering Dendrogram")
plt.show()
```

```python
# 34. Compare inertia values for K = 2 to 6

from sklearn.datasets import load_breast_cancer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

data = load_breast_cancer()
X = data.data

inertia = []
k_values = range(2,7)

for k in k_values:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

plt.plot(k_values, inertia)
plt.xlabel("K")
plt.ylabel("Inertia")
plt.show()
```
