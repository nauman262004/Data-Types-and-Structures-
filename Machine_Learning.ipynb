# Machine Learning Theory Questions

1. What is a parameter?
 - A parameter is a value that controls the behavior of a model or function. In Machine Learning, parameters are learned from data during training, like weights in linear regression. They help the model make predictions. Parameters directly affect the model's output.

2. What is correlation? What does negative correlation mean?
 - Correlation measures the relationship between two variables. It shows how one variable changes when another changes. Negative correlation means when one variable increases, the other decreases. For example, as price increases, demand may decrease.

3. Define Machine Learning. What are the main components in Machine Learning?
  - Machine Learning is a subset of AI where systems learn from data without being explicitly programmed. The main components are data, features, model, algorithm, and evaluation metrics. These components work together to make predictions or decisions.

4. How does loss value help in determining whether the model is good or not?
  - Loss value measures how far the model’s predictions are from actual values. A lower loss value means the model is performing better. During training, the goal is to minimize the loss. It helps compare different models or iterations.

5. What are continuous and categorical variables?
  - Continuous variables are numerical values that can take any value within a range, like height or weight. Categorical variables represent categories or labels, like gender or color. They are usually non-numeric. Each type is handled differently in ML.

6. How do we handle categorical variables in Machine Learning? What are the common techniques?
  - Categorical variables are converted into numerical form for ML models. Common techniques include Label Encoding and One-Hot Encoding. Label Encoding assigns numbers to categories. One-Hot Encoding creates separate columns for each category.

7. What do you mean by training and testing a dataset?
  - Training data is used to teach the model patterns from data. Testing data is used to evaluate how well the model performs on unseen data. This helps check model accuracy. It prevents overfitting.

8. What is sklearn.preprocessing?
  - sklearn.preprocessing is a module in Scikit-learn used for data preprocessing. It provides tools for scaling, encoding, and normalization. Examples include StandardScaler and LabelEncoder. It prepares data for better model performance.

9. What is a Test set?
  - A test set is a portion of data used to evaluate the trained model. It is not used during training. It checks how well the model generalizes to new data. Model accuracy is often measured on the test set.

10. How do we split data for model fitting (training and testing) in Python? How do you approach a Machine Learning problem?
 - Data is split using train_test_split() from sklearn.model_selection. Usually, 70–80% is training data and the rest is testing data. A ML problem is approached by understanding data, preprocessing, selecting a model, training, and evaluation.

11. Why do we have to perform EDA before fitting a model to the data?
 - EDA (Exploratory Data Analysis) helps understand data patterns and distributions. It identifies missing values, outliers, and relationships between variables. This improves feature selection and preprocessing. EDA leads to better model performance.

12. What is correlation?
  - Correlation shows the strength and direction of the relationship between two variables. It ranges from -1 to +1. A value close to +1 means strong positive correlation. A value close to 0 means weak or no correlation.

13. What does negative correlation mean?
  - Negative correlation means two variables move in opposite directions. When one increases, the other decreases. The correlation value lies between 0 and -1. Example: speed and travel time.

14. How can you find correlation between variables in Python?
  - Correlation can be found using the .corr() method in pandas. It calculates correlation coefficients between numerical columns. Example: df.corr(). It helps identify relationships between features.

15. What is causation? Explain difference between correlation and causation with an example.
  - Causation means one variable directly causes a change in another. Correlation only shows a relationship, not cause. Example: Ice cream sales and drowning are correlated, but ice cream does not cause drowning. Heat is the real cause.

16. What is an Optimizer? What are different types of optimizers? Explain each with an example.
  - An optimizer adjusts model parameters to minimize loss. Examples include Gradient Descent, Stochastic Gradient Descent (SGD), and Adam. SGD updates parameters using one data point at a time. Adam adapts learning rates for faster convergence.  

17. What is sklearn.linear_model?
  - sklearn.linear_model is a module in Scikit-learn. It provides linear models like Linear Regression and Logistic Regression. These models are used for prediction and classification. They are simple and efficient.

18. What does model.fit() do? What arguments must be given?
 - model.fit() trains the model on training data. It learns patterns from input features and target values. Required arguments are X_train and y_train. After fitting, the model is ready for prediction.

19. What does model.predict() do? What arguments must be given?
  - model.predict() generates predictions using the trained model. It takes input features as argument. Example: model.predict(X_test). It outputs predicted values or labels.

20. What are continuous and categorical variables?
  - Continuous variables are numeric and measurable, like temperature or salary. Categorical variables represent groups or labels, like city or department. They can be nominal or ordinal. Encoding is required for categorical data.  

21. What is feature scaling? How does it help in Machine Learning?
  - Feature scaling standardizes the range of input features. It helps algorithms converge faster and perform better. Models like KNN and gradient descent depend on distance. Common methods include normalization and standardization.

22. How do we perform scaling in Python?
  - Scaling is done using tools like StandardScaler or MinMaxScaler. These are available in sklearn.preprocessing. The scaler is first fitted on training data. Then it transforms both training and test data.

23. What is sklearn.preprocessing?
  - It is a Scikit-learn module used for data transformation. It includes scaling, encoding, and normalization tools. Examples are OneHotEncoder and StandardScaler. It prepares data for ML models.

24. How do we split data for model fitting (training and testing) in Python?
  - We use train_test_split() from Scikit-learn. It divides data into training and testing sets. Example: train_test_split(X, y, test_size=0.2). It ensures unbiased model evaluation.

25. Explain data encoding.
  - Data encoding converts categorical data into numerical form. ML models work only with numbers. Common encoding methods are Label Encoding and One-Hot Encoding. It helps models understand categorical features.
