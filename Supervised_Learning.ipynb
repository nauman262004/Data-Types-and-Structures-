# Theory Questions

1. What is Simple Linear Regression (SLR)? Explain its purpose.
  - Simple Linear Regression (SLR) is a statistical method used to model the relationship between one independent variable and one dependent variable using a straight line. Its purpose is to understand how the dependent variable changes with respect to the independent variable and to make predictions.

2. What are the key assumptions of Simple Linear Regression?
  - The key assumptions of Simple Linear Regression are linearity, independence of observations, homoscedasticity (constant variance of errors), normality of residuals, and absence of multicollinearity.

4. Provide a real-world example where simple linear regression can be applied.
  - Simple linear regression can be used to predict house prices based on the size of the house in square feet.

5. What is the method of least squares in linear regression?
  - The method of least squares is used to estimate regression coefficients by minimizing the sum of the squared differences between observed values and predicted values.

6. What is Logistic Regression? How does it differ from Linear Regression?
  - Logistic Regression is a classification algorithm used when the dependent variable is binary. Unlike Linear Regression, which predicts continuous values, Logistic Regression predicts probabilities using a sigmoid function.

7. Name and briefly describe three common evaluation metrics for regression models.
  - Three common evaluation metrics are Mean Squared Error (MSE), which measures average squared errors; Root Mean Squared Error (RMSE), which is the square root of MSE; and Mean Absolute Error (MAE), which measures average absolute errors.

8. What is the purpose of the R-squared metric in regression analysis?
  - R-squared measures the proportion of variance in the dependent variable explained by the independent variable. A higher R-squared value indicates a better model fit.  

10. How do you interpret the coefficients in a simple linear regression model?
  - The intercept represents the expected value of the dependent variable when the independent variable is zero. The slope represents the expected change in the dependent variable for a one-unit increase in the independent variable.  
  

# Practical Questions

```python
# 10. Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.

from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 5, 4, 5])

model = LinearRegression()
model.fit(X, y)

print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```
