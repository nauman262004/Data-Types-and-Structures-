# SVM & Naive bayes Theory Questions

1. What is a Support Vector Machine (SVM)?

  - A Support Vector Machine (SVM) is a supervised machine learning algorithm used for classification and regression.
It works by finding the optimal hyperplane that separates data points of different classes with the maximum margin.
SVM focuses on boundary points (support vectors) rather than all data points.

2. What is the difference between Hard Margin and Soft Margin SVM?

  - Hard Margin SVM assumes data is perfectly linearly separable and does not allow misclassification.
Soft Margin SVM allows some misclassification to handle noisy or overlapping data.
Soft margin uses a penalty parameter (C) to balance margin width and classification errors.

3. What is the mathematical intuition behind SVM?

  - SVM tries to maximize the margin between two classes while correctly classifying data points.
It solves an optimization problem to find the best separating hyperplane.
The idea is that a larger margin leads to better generalization on unseen data.

4. What is the role of Lagrange Multipliers in SVM?

  - Lagrange multipliers help convert the constrained optimization problem into an unconstrained form.
They identify which training points influence the decision boundary.
Only points with non-zero multipliers become support vectors.

5. What are Support Vectors in SVM?

  - Support vectors are the data points closest to the decision boundary.
They define the position and orientation of the hyperplane.
Removing them would change the model, while other points have little effect.

6. What is a Support Vector Classifier (SVC)?

  - Support Vector Classifier (SVC) is the classification version of SVM.
It finds the optimal boundary to separate classes in feature space.
It can use kernels to handle non-linear classification problems.

7. What is a Support Vector Regressor (SVR)?

  - Support Vector Regressor (SVR) is the regression version of SVM.
Instead of separating classes, it fits a function within a margin of tolerance (epsilon).
It minimizes prediction error while keeping the model as flat as possible.

8. What is the Kernel Trick in SVM?

  - The kernel trick maps data into a higher-dimensional space without explicit transformation.
This allows SVM to separate non-linearly separable data using linear boundaries in that space.
It makes complex classification efficient and computationally feasible.

9. Compare Linear Kernel, Polynomial Kernel, and RBF Kernel?

  - Linear kernel is used when data is linearly separable and is computationally simple.
Polynomial kernel captures curved relationships and interactions between features.
RBF (Gaussian) kernel handles highly non-linear data and is the most flexible but needs tuning.

10. What is the effect of the C parameter in SVM?

  - C controls the trade-off between margin size and classification error.
High C tries to classify all training points correctly (smaller margin, risk of overfitting).
Low C allows more misclassification (larger margin, better generalization).

11. What is the role of the Gamma parameter in RBF Kernel SVM?

 - Gamma defines how far the influence of a single training point reaches.
High gamma creates complex boundaries focusing on nearby points (overfitting risk).
Low gamma creates smoother, broader decision boundaries.

12. What is the Naïve Bayes classifier, and why is it called "Naïve"?

  - Naïve Bayes is a probabilistic classifier based on Bayes’ theorem.
It assumes features are conditionally independent given the class label.
It is called “naïve” because this independence assumption is rarely true in real data.

13. What is Bayes’ Theorem?

  - Bayes’ theorem describes how to update probability based on new evidence.
It states: Posterior = (Likelihood × Prior) / Evidence.
It is the foundation of probabilistic classification methods like Naïve Bayes.

14. Explain the differences between Gaussian, Multinomial, and Bernoulli Naïve Bayes?

 - Gaussian NB assumes continuous features with normal distribution.
Multinomial NB works with count-based discrete features (e.g., word counts).
Bernoulli NB works with binary features (presence/absence of features).

15. When should you use Gaussian Naïve Bayes over other variants?

  - Use Gaussian NB when features are continuous and approximately normally distributed.
It is suitable for real-valued measurements like sensor or numerical data.
It is not ideal for text or count-based data.

16. What are the key assumptions made by Naïve Bayes?

 - Features are conditionally independent given the class label.
Each feature contributes equally and independently to classification.
The model assumes feature distributions follow a chosen probability form.

17. What are the advantages and disadvantages of Naïve Bayes?

  - Advantages: fast, simple, works well with small datasets and high dimensions.
    Disadvantages: independence assumption unrealistic, sensitive to feature correlation.
It may perform poorly when features strongly depend on each other.

18. Why is Naïve Bayes a good choice for text classification?

 - Text data has many features (words) and sparse representation.
Naïve Bayes handles high-dimensional sparse data efficiently.
It works well with word frequency or presence features.

19. Compare SVM and Naïve Bayes for classification tasks?

 - SVM is a discriminative model focusing on decision boundaries.
Naïve Bayes is a probabilistic generative model based on feature distributions.
SVM usually gives higher accuracy; Naïve Bayes is faster and simpler.

20. How does Laplace Smoothing help in Naïve Bayes?

  - Laplace smoothing adds a small constant to all feature counts.
It prevents zero probability for unseen features in training data.
This ensures stable probability estimation and better predictions.

# SVM & Naive bayes Practical Questions

```python
# 21. Write a Python program to train an SVM Classifier on the Iris dataset and evaluate accuracy.

from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

iris = datasets.load_iris()
X, y = iris.data, iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = SVC(kernel='linear')
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# 22. Write a Python program to train two SVM classifiers with Linear and RBF kernels on the Wine dataset, then compare their accuracies.

from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

wine = load_wine()
X, y = wine.data, wine.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

linear_svm = SVC(kernel='linear')
rbf_svm = SVC(kernel='rbf')

linear_svm.fit(X_train, y_train)
rbf_svm.fit(X_train, y_train)

print("Linear Accuracy:", accuracy_score(y_test, linear_svm.predict(X_test)))
print("RBF Accuracy:", accuracy_score(y_test, rbf_svm.predict(X_test)))

```

```python
# 23. Write a Python program to train an SVM Regressor (SVR) on a housing dataset and evaluate it using Mean Squared Error (MSE).

from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error

data = fetch_california_housing()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = SVR(kernel='rbf')
model.fit(X_train, y_train)

pred = model.predict(X_test)
print("MSE:", mean_squared_error(y_test, pred))
```

```python
# 24. Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision boundary.

import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC

iris = datasets.load_iris()
X = iris.data[:, :2]
y = iris.target

model = SVC(kernel='poly', degree=3)
model.fit(X, y)

x_min, x_max = X[:,0].min()-1, X[:,0].max()+1
y_min, y_max = X[:,1].min()-1, X[:,1].max()+1
xx, yy = np.meshgrid(np.arange(x_min,x_max,0.02),
                     np.arange(y_min,y_max,0.02))

Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:,0], X[:,1], c=y)
plt.title("Polynomial SVM Decision Boundary")
plt.show()

```

```python
# 26. Write a Python program to train a Multinomial Naïve Bayes classifier for text classification using the 20 Newsgroups dataset.

from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

data = fetch_20newsgroups(subset='all')
X, y = data.data, data.target

vectorizer = CountVectorizer(stop_words='english')
X_vec = vectorizer.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_vec,y,test_size=0.3,random_state=42)

model = MultinomialNB()
model.fit(X_train, y_train)

print("Accuracy:", accuracy_score(y_test, model.predict(X_test)))
```

```python
# 27. Write a Python program to train an SVM Classifier with different C values and compare the decision boundaries visually.

import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC

iris = datasets.load_iris()
X = iris.data[:, :2]
y = iris.target

C_values = [0.1, 1, 10]

for C in C_values:
    model = SVC(kernel='linear', C=C)
    model.fit(X, y)

    xx, yy = np.meshgrid(np.linspace(X[:,0].min(), X[:,0].max(), 200),
                         np.linspace(X[:,1].min(), X[:,1].max(), 200))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

    plt.figure()
    plt.contourf(xx, yy, Z, alpha=0.3)
    plt.scatter(X[:,0], X[:,1], c=y)
    plt.title(f"SVM Decision Boundary (C={C})")

plt.show()
```

```python
# 28. Write a Python program to train a Bernoulli Naïve Bayes classifier for binary classification on a dataset with binary features.

from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import accuracy_score

X, y = make_classification(n_samples=1000, n_features=10, random_state=42)

# Convert to binary
X = (X > 0).astype(int)

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)

model = BernoulliNB()
model.fit(X_train, y_train)

print("Accuracy:", accuracy_score(y_test, model.predict(X_test)))
```

```python
# 29. Write a Python program to apply feature scaling before training an SVM model and compare results with unscaled data.

from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

data = load_wine()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)

# Without scaling
model1 = SVC()
model1.fit(X_train, y_train)
acc1 = accuracy_score(y_test, model1.predict(X_test))

# With scaling
scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)

model2 = SVC()
model2.fit(X_train_s, y_train)
acc2 = accuracy_score(y_test, model2.predict(X_test_s))

print("Without Scaling:", acc1)
print("With Scaling:", acc2)
```

```python
# 30. Write a Python program to train a Gaussian Naïve Bayes model and compare the predictions before and after Laplace Smoothing.

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

data = load_breast_cancer()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)

# Default
model1 = GaussianNB()
model1.fit(X_train, y_train)
acc1 = accuracy_score(y_test, model1.predict(X_test))

# With smoothing (var_smoothing)
model2 = GaussianNB(var_smoothing=1e-2)
model2.fit(X_train, y_train)
acc2 = accuracy_score(y_test, model2.predict(X_test))

print("Before smoothing:", acc1)
print("After smoothing:", acc2)
```

```python
# 31. Write a Python program to train an SVM Classifier and use GridSearchCV to tune the hyperparameters (C, gamma, kernel).

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC

data = load_iris()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)

param_grid = {
    'C':[0.1,1,10],
    'gamma':[0.01,0.1,1],
    'kernel':['linear','rbf']
}

grid = GridSearchCV(SVC(), param_grid, cv=5)
grid.fit(X_train, y_train)

print("Best Params:", grid.best_params_)
print("Best Accuracy:", grid.score(X_test, y_test))
```

```python
# 32. Write a Python program to train an SVM Classifier on an imbalanced dataset and apply class weighting and check it improve accuracy.

from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

X, y = make_classification(n_samples=1000, weights=[0.9,0.1], random_state=42)

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)

# Without weighting
model1 = SVC()
model1.fit(X_train, y_train)
acc1 = accuracy_score(y_test, model1.predict(X_test))

# With class weighting
model2 = SVC(class_weight='balanced')
model2.fit(X_train, y_train)
acc2 = accuracy_score(y_test, model2.predict(X_test))

print("Without weighting:", acc1)
print("With weighting:", acc2)
```
